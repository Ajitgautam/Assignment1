```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
HEADERS = ({'User-Agent':  'Chrome/90.0.4430.212 Safari/537.36',
            'Accept-Language': 'en-US, en;q=0.5'})
```

```python
url = "https://www.amazon.in/s?k=bags&crid=2M096C61O4MLT&qid=1653308124&sprefix=ba%2Caps%2C283&ref=sr_pg_1"
```

```python
r = requests.get(url, headers=HEADERS)
```

```python
soup = BeautifulSoup(r.content,"html.parser")
```

```python
soup
```

```python
links = soup.find_all("a", attrs = {'class' :'a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal'})
```

```python
links
```

```python
all_links = []
```

```python
for link in links:
    all_links.append(link.get('href'))
```

```python
def get_description(soup):
  try:
    description = soup.find("span", attrs={"id":'productTitle'})
    description_value = description.text
    description_string = description_value.strip()
  except AttributeError:
        description_string = ""
  return description_string
```

```python
def get_ASIN(soup):
  try:
    asin = soup.find('span', attrs={'class': 'a-list-item'}).string.strip()
  except AttributeError:
    asin = ""
    return asin
```

```python
def get_product_description(soup):
  try:
    product = soup.find('div',attrs={'id': 'productDescription'} ).string.strip()
  except AttributeError:
    product = "Not Available"
    return product
```

```python
def get_manufacturer(soup):
  try:
    Manufacturer = soup.find('span', attrs= {'class': 'a-text-bold'}).string.strip()
  except AttributeError:
        Manufacturer = ""
  return Manufacturer
```

```python
d = {"Description":[], "ASIN":[], "Product_description":[], "Manufacturer":[]}
```

```python
for link in all_links:
        new_webpage = requests.get("https://www.amazon.com" + link, headers=HEADERS)
        new_soup = BeautifulSoup(new_webpage.content, 'html.parser')
        d['Description'].append(get_description(new_soup))
        d['ASIN'].append(get_ASIN(new_soup))
        d['Product_description'].append(get_product_description(new_soup))
        d['Manufacturer'].append(get_manufacturer(new_soup))



```

```python
amazon_df = pd.DataFrame.from_dict(d)

amazon_df.to_csv("amazon_data.csv", header=True, index=False)
```
